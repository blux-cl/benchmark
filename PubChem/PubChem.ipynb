```python
import pubchempy as pcp
import pandas as pd
```

```python
compound_names = ['glucose', 'sucrose', 'D-glucitol', 'mannitol', 'fructose',
'acetic acid', 'glutamic acid', 'citric acid', 'L-ascorbic acid', 'benzoic acid',
'pectin', 'carboxymethylcellulose', 'gum arabic', 'xanthan',
'Maltodextrin-dextrose equivalent 10-15', 'fructooligosaccharide', 'inulin', 'sucralose',
'aspartame', 'stevioside', 'stevia', 'phosphoric acid',
'sodium carbonate', 'Dioleoyl lecithin', 'alginate', 'carrageenan',
'agar', 'Amylum', 'Papain Inhibitor', 'gallic acid',
'potassium chloride', 'carboxymethylcellulose', 'calcium nitrate']
```

# Compound
Retrieving information about a specific Compound in the PubChem database is simple.

```python
c = pcp.Compound.from_cid(5793)
df = pcp.compounds_to_frame(c)
columns = list(df.columns)
columns.insert(0, 'name')
columns.insert(0, 'cid')
df
```

```python
columns
```

```python
df = pd.DataFrame([], columns = columns)
df
```

```python
for compound_name in compound_names:
    c =  pcp.get_compounds(compound_name, 'name')
    if len(c) != 0:
        c = c[0]
        print(f"{compound_name}:{c}")
        d = pcp.compounds_to_frame(c)
        d.insert(0, "name", compound_name)
        d.insert(0, "cid", c.cid)
        df = pd.concat([df, d])
    #else:
    #    print(f"{compound_name}:{len(c)}, {c}")
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
df.to_csv('pubchem.csv', index=False)
```
